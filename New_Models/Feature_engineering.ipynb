```python
from Feature_engineering import *
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

'''normalizes values to be within 0 and max_val'''
def normalize_non_label_columns_min_max(df, all_labels, max_val):
    """
    Normalizes (using Min-Max scaling) all columns in the dataframe except those listed in all_labels.

    Args:
    df (pandas.DataFrame): The dataframe to be normalized.
    all_labels (list): List of column names not to be normalized.

    Returns:
    pandas.DataFrame: A new dataframe with the non-label columns normalized.
    """
    # Select columns that are not in all_labels
    columns_to_normalize = df.columns.difference(all_labels)

    # Normalize these columns using Min-Max scaling
    scaler = MinMaxScaler()
    df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize]) * max_val

    return df

'''normalizes values with the StandardScaler, which removes the mean and then scales based on variance.'''
def normalize_non_label_columns(df, all_labels):
    """
    Normalizes all columns in the dataframe except those listed in all_labels.

    Args:
    df (pandas.DataFrame): The dataframe to be normalized.
    all_labels (list): List of column names not to be normalized.

    Returns:
    pandas.DataFrame: A new dataframe with the non-label columns normalized.
    """
    # Select columns that are not in all_labels
    columns_to_normalize = df.columns.difference(all_labels)

    # Normalize these columns
    scaler = StandardScaler()
    df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

    return df
```
